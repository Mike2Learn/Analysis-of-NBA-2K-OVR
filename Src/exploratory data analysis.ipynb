```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats as STATS
```

```python
path = "C:/Users/74279/PycharmProjects/NBA2K-Ratings/Data/"
```

```python
stats = pd.read_csv(path+'2015-2022 NBA Player Stats.csv')
stats = stats.iloc[:,1:]
stats = stats[stats.GP > 41]
```

```python
#Checking whether the response is normal
#from STATS import norm
log_rating = np.log1p(stats["RATING"])
sns.distplot(log_rating,fit=STATS.norm);
fig = plt.figure()
res = STATS.probplot(log_rating, plot=plt)
```
