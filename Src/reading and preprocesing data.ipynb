```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl
import xlrd
```

```python
path = "C:/Users/74279/PycharmProjects/NBA2K-Ratings/Data/"
path
```

```python
#Read in 2021-2022 NBA Player Stats
stats_2122 = pd.read_excel(path+"2021-2022 NBA Player Stats.xlsx")
stats_2122.columns = stats_2122.iloc[0]
stats_2122 = stats_2122.iloc[1:,1:]

#Read in NBA 2k23 Ratings
ratings_23 = pd.read_excel(path+"2k23 ratings.xlsx")

#Merge these two dataframes
stats_2122 = stats_2122.merge(ratings_23,left_on="FULL NAME",right_on="FULL NAME")
```

```python
stats_2122
```

```python
#Read in 2020-2021 NBA Player Stats
stats_2021 = pd.read_excel(path+"2020-2021 NBA Player Stats.xlsx")
stats_2021.columns = stats_2021.iloc[0]
stats_2021 = stats_2021.iloc[1:,1:]

#Read in NBA 2k22 Ratings
ratings_22 = pd.read_excel(path+"2k22 ratings.xlsx")

#Merge these two dataframes
stats_2021 = stats_2021.merge(ratings_22,left_on="FULL NAME",right_on="FULL NAME")
```

```python
stats_2021
```

```python
#Read in 2021-2022 NBA Player Stats
stats_1920 = pd.read_excel(path+"2019-2020 NBA Player Stats.xlsx")
stats_1920.columns = stats_1920.iloc[0]
stats_1920 = stats_1920.iloc[1:,1:]

#Read in NBA 2k21 Ratings
ratings_21 = pd.read_excel(path+"2k21 ratings.xlsx")
#Merge these two dataframes
stats_1920 = stats_1920.merge(ratings_21,left_on="FULL NAME",right_on="FULL NAME")
```

```python
stats_1920
```

```python
#Read in 2018-2019 NBA Player Stats
stats_1819 = pd.read_excel(path+"2018-2019 NBA Player Stats.xlsx")
stats_1819.columns = stats_1819.iloc[0]
stats_1819 = stats_1819.iloc[1:,1:]

#Read in NBA 2k20 Ratings
ratings_20 = pd.read_excel(path+"2k20 ratings.xlsx")
#Merge these two dataframes
stats_1819 = stats_1819.merge(ratings_20,left_on="FULL NAME",right_on="FULL NAME")
```

```python
stats_1819
```

```python
#Read in 2017-2018 NBA Player Stats
stats_1718 = pd.read_excel(path+"2017-2018 NBA Player Stats.xlsx")
stats_1718.columns = stats_1718.iloc[0]
stats_1718 = stats_1718.iloc[1:,1:]

#Read in NBA 2k20 Ratings
ratings_19 = pd.read_excel(path+"2k19 ratings.xlsx")
#Merge these two dataframes
stats_1718 = stats_1718.merge(ratings_19,left_on="FULL NAME",right_on="FULL NAME")
#Uniform column names
stats_1718.columns = stats_1819.columns
```

```python
stats_1718
```

```python
#Combine these stats and drop columns with nan
frames = [stats_2122,stats_2021,stats_1920,stats_1819,stats_1718]
stats_1722 = pd.concat(frames)
stats_1722 = stats_1722.dropna(axis=1)
```

```python
#The column names are too prolix, so let's make it shorter
stats_1722.rename(columns=lambda x: x[:3], inplace=True)
stats_1722.rename(columns={"FUL":"FULL NAME","TEA":"TEAM","RAT":"RATING"},inplace=True)
#Drop column 'TOP'
stats_1722 = stats_1722.drop("TOP",axis=1)
```

```python
stats_1722[stats_1722["FULL NAME"]=="James Harden"]
```

```python
#Read in 2016-2017 NBA Player Stats
stats_1617 = pd.read_excel(path+"2016-2017 NBA Player Stats.xlsx")
stats_1617.columns = stats_1617.iloc[0]
stats_1617 = stats_1617.iloc[1:,1:]

#Read in NBA 2k18 Ratings
ratings_18 = pd.read_excel(path+"2k18 ratings.xlsx")
#Merge these two dataframes
stats_1617 = stats_1617.merge(ratings_18,left_on="FULL NAME",right_on="FULL NAME")
#Add and drop some column
stats_1617 = stats_1617.drop('TOr',axis=1)
stats_1617 = stats_1617.drop('TS%',axis=1)
#Uniform column names
stats_1617.columns = stats_1722.columns
```

```python
stats_1617[stats_1617["FULL NAME"]=="James Harden"]
```

```python
#Read in 2015-2016 NBA Player Stats
stats_1516 = pd.read_excel(path+"2015-2016 NBA Player Stats.xlsx")
stats_1516.columns = stats_1516.iloc[0]
stats_1516 = stats_1516.iloc[1:,1:]

#Read in NBA 2k17 Ratings
ratings_17 = pd.read_excel(path+"2k17 ratings.xlsx")
#Merge these two dataframes
stats_1516 = stats_1516.merge(ratings_17,left_on="FULL NAME",right_on="FULL NAME")
#Add and drop some column
stats_1516 = stats_1516.drop('TOr',axis=1)
stats_1516 = stats_1516.drop('TS%',axis=1)
#Uniform column names
stats_1516.columns = stats_1722.columns
```

```python
stats_1516[stats_1516["FULL NAME"]=="James Harden"]
```

```python
stats_1522 = pd.concat([stats_1722,stats_1617,stats_1516])
stats_1522 = stats_1522.dropna(axis=1)
```

```python
stats_1522
```

```python
stats_1522[stats_1522["FULL NAME"]=="James Harden"]
```
