```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
```

```python
stats = pd.read_csv('/content/2015-2022 NBA Player Stats.csv')
stats = stats.iloc[:,1:]
stats = stats[stats.GP > 41]
```

```python
x = stats.iloc[:,5:22]
x = x.drop(["MPG","MIN"],axis=1)
y = stats.iloc[:,-1:]
from sklearn.model_selection import train_test_split
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error as mae
scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=671)

MLPreg = MLPRegressor(random_state=1, max_iter=1000)
MLPreg.fit(x_train,y_train)
print(mae(MLPreg.predict(x_train),y_train))
print(mae(MLPreg.predict(x_test),y_test))
```

```python
param_grid = {'hidden_layer_sizes': [(100,50,50), (100,50,100), (100,)],
          'activation': ['relu','tanh','logistic','sigmoid'],
          'alpha': [0.001, 0.05],
          'learning_rate': ['constant','adaptive'],
          'solver': ['adam']}
```

```python
from sklearn.model_selection import GridSearchCV
mlp = MLPRegressor()
grid_search = GridSearchCV(estimator=mlp,param_grid = param_grid,cv=5, scoring='neg_mean_squared_error', verbose=0, n_jobs=-1)
```

```python
grid_search.fit(x_train,y_train)
grid_search.best_params_
```

```python
best_grid = grid_search.best_estimator_
print(mae(best_grid.predict(x_train),y_train))
print(mae(best_grid.predict(x_test),y_test))
```
